```python
#Import Dependencies
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
```

```python
#Import csv File using pandas
anscombe_df = pd.read_csv('/content/drive/MyDrive/SGA_csvdocs/anscombe_quartet_data.csv')
anscombe_df
```

```python
#Get the statistics
anscombe_df.describe()
```

```python
#Correlation Data of anscombe_df
anscombe_df.corr()
```

```python
#How x correlates with y using scatterplot and line plot in a subplot
#Implementing the Four methods of creating subplots to visualize the data
#First Method
fig = plt.figure(figsize=(8, 6))
x = [0,14]
y = [0,10.84]
x1 = anscombe_df['x1']
y1 = anscombe_df['y1']
plt.subplot(2, 2, 1)
plt.title('(x1,y1)')
plt.xlabel('x1')
plt.ylabel('y1')
plt.scatter(x1, y1, color ="yellow")
plt.plot(x,y, color = 'black', linestyle = '-')

x = [0,14]
y = [0,9.26]
x2 = anscombe_df['x2']
y2 = anscombe_df['y2']
plt.subplot(2, 2, 2)
plt.title('(x2,y2)')
plt.xlabel('x2')
plt.ylabel('y2')
plt.scatter(x2, y2, color ="blue")
plt.plot(x,y, color = 'black', linestyle = '-')

x = [0,14]
y = [0,12.74]
x3 = anscombe_df['x3']
y3 = anscombe_df['y3']
plt.subplot(2, 2, 3)
plt.title('(x3,y3)')
plt.xlabel('x3')
plt.ylabel('y3')
plt.scatter(x3, y3, color ="purple")
plt.plot(x,y, color = 'black', linestyle = '-')

x = [0,19]
y = [0,12.5]
x4 = anscombe_df['x4']
y4 = anscombe_df['y4']
plt.subplot(2, 2, 4)
plt.title('(x4,y4)')
plt.xlabel('x4')
plt.ylabel('y4')
plt.scatter(x4, y4, color ="green")
plt.plot(x,y, color = 'black', linestyle = '-')

plt.subplots_adjust(wspace=0.5, hspace=0.5, bottom=0.1, top=0.9)
plt.show()
```

```python
#Second Method
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(8, 6))

x = [0,14]
y = [0,10.84]
ax1.scatter(x1, y1, color ="yellow")
ax1.plot(x, y, color='black')
ax1.set_title('(x1,y1)')
ax1.set_xlabel('x1')
ax1.set_ylabel('y1')

x = [0,14]
y = [0,9.26]
ax2.scatter(x2, y2, color ="blue")
ax2.plot(x, y, color='black')
ax2.set_title('(x2,y2)')
ax2.set_xlabel('x2')
ax2.set_ylabel('y2')


x = [0,14]
y = [0,12.74]
ax3.scatter(x3, y3, color ="purple")
ax3.plot(x, y, color='black')
ax3.set_title('(x3,y3)')
ax3.set_xlabel('x3')
ax3.set_ylabel('y3')


x = [0,19]
y = [0,12.5]
ax4.scatter(x4, y4, color ="green")
ax4.plot(x, y, color='black')
ax4.set_title('(x4,y4)')
ax4.set_xlabel('x4')
ax4.set_ylabel('y4')
plt.subplots_adjust(wspace=0.5, hspace=0.5, bottom=0.1, top=0.9)
```

```python
#Third Method
fig = plt.figure(figsize=(8, 6))

x = [0,14]
y = [0,10.84]
ax5 = fig.add_subplot(221, title='(x1,y1)', xlabel='x1', ylabel='y1')
ax5.scatter(x1, y1, c ="yellow")
ax5.plot(x, y, color='black')

x = [0,14]
y = [0,9.26]
ax6 = fig.add_subplot(222, title='(x2,y2)', xlabel='x2', ylabel='y2')
ax6.scatter(x2, y2, c ="blue")
ax6.plot(x, y, color='black')

x = [0,14]
y = [0,12.74]
ax7 = fig.add_subplot(223, title='(x3,y3)', xlabel='x3', ylabel='y3')
ax7.scatter(x3, y3, c ="purple")
ax7.plot(x, y, color='black')

x = [0,19]
y = [0,12.5]
ax8 = fig.add_subplot(224, title='(x4,y4)', xlabel='x4', ylabel='y4')
ax8.scatter(x4, y4, c ="green")
ax8.plot(x, y, color='black')

plt.subplots_adjust(wspace=0.5, hspace=0.5, bottom=0.1, top=0.9)
```
