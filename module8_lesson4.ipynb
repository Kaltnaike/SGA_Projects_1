```python
#Import Dependencies
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#load csv File into DataFrame
df = pd.read_csv('/content/drive/MyDrive/SGA_csvdocs/WorldCupMatches.csv')
df.head(5)
```

```python
#Inspecting and cleaning of Dataset
#Check Datatypes
df.dtypes
```

```python
#Inspecting and cleaning of Dataset
#Check missing values and dropping them
df.isnull().sum()
df.dropna( inplace = True)
df.head(5)
```

```python
#Create a new column in df named Total Goals
df['Total Goals'] = df['Home Team Goals'] + df['Away Team Goals']
df.head(5)
```

```python
#Visualize Total number of goals scored in each match
#fig = plt.figure(figsize=(15, 5))
sns.set(rc={'figure.figsize':(15,8)})
sns.displot(df['Total Goals'])
plt.title('Total Number of Goals scored in each match'.upper())
plt.show()
```

```python
#Create a bar chart visualizing how many goals were scored each year the World Cup was held between 1930-2014. 
fig = plt.figure(figsize=(15, 5))
sns.set_style('whitegrid')
sns.set_context(rc={'font.size': 10, 'xtick.labelsize':8})
sns.barplot(data=df, x='Year', y='Total Goals',ci=None,estimator=sum)
plt.xticks(rotation='vertical')
plt.title('Number of Goals Scored each year(1930-2014)'.upper())
plt.show()
```

```python
#Create a figure and axes for your plot, inside of plt.subplots(), set the size of the figure to be 12 inches wide and 7 inches tall.
#Visualize the columns Year and Total Goals as a bar chart.
fig= plt.subplots(figsize=(12,7))
ax = sns.barplot(data=df, x="Year", y="Total Goals", ci=None, estimator=sum)
ax.set_title("Total Goals Scored Yearly".upper())  
plt.xticks(rotation='vertical')
plt.show()
```

```python
# Create a box plot to visualize the distribution of the average goals for each year with a bar chart.
fig= plt.figure(figsize=(12,7))
sns.boxplot(data=df,x='Year', y='Total Goals')
plt.title('Average goals for each year'.upper())
plt.xticks(rotation='vertical')
plt.show()
```
