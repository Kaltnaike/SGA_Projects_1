```python
#Import Dependencies
import numpy as np 
import pandas as pd
```

```python
#Load csv File
df=pd.read_csv('/content/drive/MyDrive/SGA_csvdocs/movies.csv')
df.head(5)
```

```python
#Check for Missing Values
df.isnull().any().any()
```

```python
#Check for Missing Values in each column of the Dataset
df.isnull().any()
```

```python
#Check for Missing Values in each column of the Dataset
df.isnull().sum()
```

```python
#Check for Duplicates
df.duplicated()
```

```python
df.drop_duplicates()
df.head(5)
```

```python
#Ensure correct data types of columns
df.dtypes
```

```python
#Working on 'VOTES' datatype
#Removing Comma from 'VOTES' column
df['VOTES'] = df['VOTES'].astype(str)
df['VOTES'] = [x.replace(',','') for x in df["VOTES"]]
# convert "VOTES" from string to float
df['VOTES'] = df['VOTES'].astype(float)
#correct data types of columns
df.dtypes
```

```python
#More Data cleaning to make the dataset neat
#Removing the next line(\n) value from the columns
df["GENRE"] = df["GENRE"].astype(str)
df["GENRE"] = [x.replace('\n','') for x in df["GENRE"]]


df["ONE-LINE"] = df["ONE-LINE"].astype(str)
df["ONE-LINE"] = [x.replace('\n','') for x in df["ONE-LINE"]]

df["STARS"] = df["STARS"].astype(str)
df["STARS"] = [x.replace('\n','') for x in df["STARS"]]
df.head(5)
```

```python
#Using fillna to fill up empty spaces in the columns
# Fill up empty spaces  for 'RunTime' column using the mean
x = df['RunTime'].mean()
df['RunTime'].fillna(x, inplace=True)

# Fill up empty spaces  for 'RATING' column using the mean
y = df['RATING'].mean()
df['RATING'].fillna(y, inplace=True)

# Fill up empty spaces  for 'VOTES' column with 0
df['VOTES'].fillna(0, inplace=True)

# Fill up empty spaces  for 'Gross' column with 0
df['Gross'].fillna(0, inplace=True)

# Fill up empty spaces for 'GENRE' column with 'Comedy'
df['GENRE'].fillna('Comedy', inplace=True)

df.head(5)
```

```python
#Checking for more duplicates and removing them
df.duplicated().sum()
df.drop_duplicates(inplace= True)
df.duplicated().sum()
```

```python
#Checking for missing values
df.isnull().sum()
#dropping rows that still have missing values
df.dropna( inplace = True)
#check
df.isnull().sum()
```

```python
cleaned_df=df
print('Dataset is Cleaned and Fit for Analysis')
cleaned_df.head(10)
```
