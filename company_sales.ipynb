```python
#Import Dependecies
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#Read the file using pandas
sales_df = pd.read_csv('/content/drive/MyDrive/SGA_csvdocs/company_sales_data.csv')
sales_df
```

```python
#Compare all the products sales data for every month using the multiple line plot
sales_df.plot(x='month_number', y=['facecream', 'facewash', 'toothpaste', 'bathingsoap','shampoo', 'moisturizer'], figsize = (15,5))
#Customize lines and include marker of choice
sales_df.plot(x='month_number',y =['facecream', 'facewash', 'toothpaste', 'bathingsoap','shampoo', 'moisturizer'] ,ls='--', color=['yellow','green','blue','purple','black','red'],  marker= 'o', figsize= (20,10))
#Put the legend at the top left corner of the axes
plt.legend(loc = 'upper left')
#Label the axes properly
plt.title("Aloecare Products Sales Data By Month")
plt.xlabel('Months')
plt.ylabel("Sales of Aloecare products")
#Display the results
plt.show()
```
