```python
#Import Dependecies
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#Read the file using pandas
sales_df = pd.read_csv('/content/drive/MyDrive/SGA_csvdocs/company_sales_data.csv')
sales_df
```

```python
#Compare all the products sales data for every month using the multiple line plot
#Customize lines and include marker of choice
x = sales_df['month_number']
y1 = sales_df['facecream']
plt.plot(x,y1, color = 'black', linestyle = '--', marker = 'o')

x = sales_df['month_number']
y2 = sales_df['toothpaste']
plt.plot(x,y2, color = 'blue', linestyle = '-', marker = '+')

x = sales_df['month_number']
y3 = sales_df['facewash']
plt.plot(x,y3, color = 'green', linestyle = ':')

x = sales_df['month_number']
y4 = sales_df['bathingsoap']
plt.plot(x,y4, color = 'purple', linestyle = 'solid', marker = 'o')

x = sales_df['month_number']
y5 = sales_df['shampoo']
plt.plot(x,y5, color = 'red', linestyle = '-', marker = '+')

x = sales_df['month_number']
y6 = sales_df['moisturizer']
plt.plot(x,y6, color = 'violet', linestyle = 'None', marker = '*')

#Put the legend at the top left corner of the axes
plt.legend(['facecream', 'toothpaste', 'facewash', 'bathingsoap', 'shampoo', 'moisturizer'],loc = 'upper left')
#Label the axes properly
plt.title("Aloecare Products Monthly Sales")
plt.xlabel('Months')
plt.ylabel("Sales of Aloecare products")
#Display the results
plt.show()
```
