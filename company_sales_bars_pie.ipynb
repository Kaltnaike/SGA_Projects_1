```python
#!pip install --upgrade matplotlib to be able put figures on the bars to identify

#Import Dependencies
import numpy as np
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
```

```python
#Import csv File
sales_df = pd.read_csv('/content/drive/MyDrive/SGA_csvdocs/company_sales_data.csv')
sales_df
```

```python
#Compare any two products sales data using a bar chart.
x = np.arange(len(sales_df['month_number']))
month = ['Jan', 'Feb', 'Mar', 'April', 'May', 'June', 'July', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec']

ax = sales_df.plot.bar(x = 'month_number', y=['shampoo', 'moisturizer'], figsize = (15,10),  width = 0.7, color=['brown', 'purple'])
#To show numbers on the bars
for container in ax.containers:
    ax.bar_label(container)
plt.legend(loc = 'upper left')
plt.title("Comparison of Shampoo and Moisturizer in Aloecare")
plt.xlabel('Months')
plt.ylabel("Sales of Shampoo and Moisturizer")
plt.xticks(x, month)
plt.savefig('compare_products_sales.png')
plt.show()
```

```python
#Assuming the average monthly sales of bathing soap is 10k, Visualize the sales data of bathing soap of all months, showing how they differ from the average sale.
fig = plt.figure(figsize=(15, 5))
x = np.arange(len(sales_df['month_number']))
month = ['Jan', 'Feb', 'Mar', 'April', 'May', 'June', 'July', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec']
yerr = np.array(list(map(lambda x: 10000-x, sales_df['bathingsoap'])))
yerr = np.absolute(yerr)

y1 = sales_df['bathingsoap']
plt.bar(x,y1, yerr=yerr, capsize=10, label = 'bathingsoap',color='violet')

plt.title('Monthly Sales of Bathing Soap in Aloecare')
plt.xlabel('Months')
plt.ylabel("Total-to-Average Monthly sales")
plt.xticks(x, month)
plt.legend()
plt.savefig('Total-to-Average Monthly sales.png')
plt.show()
```

```python
#Calculate the total sales data for the year of  each product and represent it on a pie chart
fig = plt.gcf()
fig.set_size_inches(14,10) 
products_categories = ['Facecream', 'Facewash', 'Toothpaste', 'Bathingsoap','Shampoo','Moisturizer']
products_data = [sales_df['facecream'].sum(), sales_df['facewash'].sum(), sales_df['toothpaste'].sum(), 
                sales_df['bathingsoap'].sum(), sales_df['shampoo'].sum(), sales_df['moisturizer'].sum()]
plt.pie(products_data, labels=products_categories,autopct='%.d %%', colors= ['Purple','brown','pink', 'salmon','violet', 'red'])
plt.title('Total sales of Aloecare Products')
plt.axis('equal')
plt.legend(loc = 'upper left')
plt.savefig('Total sales of each Product.png')
plt.show()
```
