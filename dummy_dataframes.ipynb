```python
# Do an inner merge operation between the orders table and products table
df_orders_products = df_order.merge(df_products, how='inner', on='product_id')
df_orders_products.head()
```

The inner merge combines the columns elements that are common between the orders and products table



```python
#Use the resulting table to perform another inner merge operation with the customers table.
df_all = df_orders_products.merge(df_customers, how='inner')
df_all.head()
```

From the table it is observed that the dataframe returns rows that have common/similar properties
Pandas uses inner merge by default. This keeps only the common values in both the left and right dataframes for the merged data.

```python
#Perform an inner merge operation between customers and orders tables, will be merging customer_id
df_customers_orders = df_customers.merge(df_order, how='inner', on='customer_id')
df_customers_orders.head()
```

Only rows contained in the Recent DataFrame are the ones where the customer_id appears in df_orders and df_customers DataFrames.

```python
#Perform an outer merge between customers and products tables.
df_customers_products = pd.merge(df_customers,df_products,how ='outer')
df_customers_products
```

MergeError:No common columns to merge on


```python
#Perform an outer merge between customers and products tables. using left_index,right_index
df_customers_products = pd.merge(df_customers,df_products,left_index=True,right_index=True,how ='outer')
df_customers_products.head()
```

The customer and the product Dataframes had no row that matches. 
but after the merge both dataframes were able to join together using the matching index of the DataFrames.

It was also observed that the customer table has NaN value in some areas. This is due to the fact that Products table has more rows than the Customers Table thereby returning the NaN value in those places.


```python
#Perform a left merge between orders and customers tables. picking customer_id for merge
df_orders_customers = pd.merge(df_order,df_customers, how='left', on=['customer_id'])
df_orders_customers
```

Tables were aligned using customer_id for merging
The merged dataframe still has NaN values in some areas, the missing values are not matching/corresponding with some data in the Customers table
