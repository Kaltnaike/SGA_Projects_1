```python
#Import Dependencies
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# Load Tips Dataset
tips = sns.load_dataset('tips')
tips
```

```python
plt.figure(figsize=(15,10)) 
#Generate a KDE plot for the size feature of the tips dataset. 
plt.subplot(2,2,1)
sns.kdeplot(tips['size'], shade=True)
plt.title('KDE plot for size feature(Tips Dataset)'.upper())

#Using the boxplots, generate the distribution of total bill by gender for the tips dataset. 
plt.subplot(2,2,2)
sns.boxplot(data=tips, x='sex', y='total_bill')
plt.title('distribution of total bill by gender(Tips Dataset)'.upper())

#Using boxplots generate the distribution of tips by days.
plt.subplot(2,2,3)
sns.boxplot(data=tips, x='day', y='tip')
plt.title('distribution of tips by days(Tips Dataset)'.upper())

#Using boxplots generate the distribution of total bill by days considering smoker status of the customer.
plt.subplot(2,2,4)
sns.boxplot(data=tips, x='day', y='total_bill', hue='smoker')
plt.title('distribution of total bill by days in \nsmoker status of customer(Tips Dataset)'.upper())

plt.subplots_adjust(wspace=0.5, hspace=0.5, bottom=0.1, top=0.9)
plt.show()
```

**Observations of the Tips Dataset Subplots**


*   Subplot 1(KDE plots for Size Feature)- Skewness is to the right, mode of the size is at 2, and it is multi-modal at size 3,4,6.

*   Subplot 2(Distribution of Total bill by gender)- The inter-quartile range in Male is higher in Female. It is observed they have the almost or the same number of outliers but difference in the spacing.
In Conclusion,Male pay more bills than female.

*   Subplot 3(Distribution of Tips By days)- Highest Inter-quartile range is at Sunday,Outliers are found in Thursday and Saturday but Saturday has more outliers. 
*   Subplot 4(Distribution of Total Bills By days in Smoker Status of Customer)- Smoking status is not a major factor in the Dataset.Smokers buy more on Saturday and Sunday.












```python
plt.figure(figsize=(15,10))

#Using violin plots generate the distribution of total bill by time given the genders.
plt.subplot(2,2,1)
sns.violinplot(data=tips,  x=tips['total_bill'],y=tips['time'], hue='sex', orient='horizontal')
plt.title('distribution of total bill by time in genders\n(tips dataset)'.upper())

#Using violin plots generate the distribution of tips by days also considering time.
plt.subplot(2,2,2)
sns.violinplot(data=tips, y=tips['day'], x=tips['tip'], hue='time', orient='horizontal')
plt.title('distribution of tips by days in time(tips dataset)'.upper())

plt.subplots_adjust(wspace=0.5, hspace=0.5, bottom=0.1, top=0.9)
plt.show()
```

**Observations of the Tips Dataset Subplots**

*   Subplot 1(Distribution of Total bill by Time in Genders)- More Sales at Dinner than lunch,during dinner male has more upper outliers male frequent more at dinner than female, female violin plot for lunch is more peaked than the male violin plot at lunch,female frequent more at lunch than male.
*   Subplot 2(Distribution of tips by days in time)- Tipping happens more during dinner, except on Thurday where tipping at lunch is higher than tipping at dinner.



```python
#Plot and customize a correlations heatmap for the tips dataset.
corr=tips.corr()
# use numpy function to mask the data to create a matrix of booleans
ones = np.ones_like(corr, dtype='bool') # by default returns ones for all the data

# use another numpy function to get just the upper triangle of the boolean array
mask = np.triu(ones)

# adjust mask and correlations 
mask = mask[1:, :-1]
corr = corr.iloc[1:,:-1].copy()

# generate customized heatmap of correlations
fig, ax = plt.subplots(figsize=(8,6))
sns.heatmap(corr, mask=mask, annot=True,fmt=".1f",cmap='Blues', linewidths=.9, cbar_kws={'shrink':.99})
ax.xaxis.tick_bottom()
plt.title("Correlations heatmap of Tips Dataset \n".upper(), fontdict={'fontsize': 18}, loc='left')
plt.show()

```
